```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
import requests  # 외부에 있는 이미지를 가져올 때 사용
from io import BytesIO

url = 'https://cdn.pixabay.com/photo/2018/10/01/09/21/pets-3715733_960_720.jpg'

response = requests.get(url)
pic = Image.open(BytesIO(response.content))

pic_arr = np.asarray(pic)
```

RGB 분포 확인

      - R channel 분포 확인

```python
# Red를 제외한 Blue 와 green의 값을 다 0으로 바꿈
pic_red = pic_arr.copy()
pic_red[:, :, 1] = 0
pic_red[:, :, 2] = 0

pic_red
```

```python
plt.imshow(pic_red)
plt.show()
```

      G channel 확인

```python
# green를 제외한 blue 와 red의 값을 다 0으로 바꿈
pic_green = pic_arr.copy()
pic_green[:, :, 0] = 0
pic_green[:, :, 2] = 0

pic_green
```

```python
plt.imshow(pic_green)
plt.show()
```

      B channel 확인

```python
# green를 제외한 blue 와 red의 값을 다 0으로 바꿈
pic_blue = pic_arr.copy()
pic_blue[:, :, 0] = 0
pic_blue[:, :, 1] = 0

pic_blue
```

```python
plt.imshow(pic_blue)
plt.show()
```
