```python
import cv2
import numpy
import matplotlib.pyplot as plt
```

```python
pic = cv2.imread('dogs.bmp')
```

```python
type(pic)
```

```python
cv2.imshow('dog', pic)
cv2.waitKey(0)             
cv2.destroyAllWindows()  
```

```python
pic.shape
```

```python
pic = cv2.cvtColor(pic, cv2.COLOR_BGR2RGB)
plt.imshow(pic)
```

```python
scale = 0.5
new_width = int(pic.shape[1]*scale)
new_height = int(pic.shape[0]*scale)
```

```python
pic_resize = cv2.resize(pic, (new_width, new_height))
pic_resize.shape
```

```python
plt.imshow(pic_resize)
```
